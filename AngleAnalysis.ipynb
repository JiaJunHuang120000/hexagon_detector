```python
import numpy as np, uproot as ur, awkward as ak, pandas as pd, uproot as ur
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from scipy.interpolate import CubicSpline
from scipy.optimize import curve_fit

import mplhep as hep
plt.figure()
hep.style.use("CMS")

import uproot,glob,ROOT

def gaus(x, amp, mean, sigma):
    return amp * np.exp( -(x - mean)**2 / (2*sigma**2) ) 
```

```python
arrays={}
arrays['muon']=ur.open(f"data/muon_4GeV_Theta_90_150_Phi_0_360_Events_50000_Hexagon.edm4hep.root:events").arrays()
```

```python
px = arrays['muon']["MCParticles.momentum.x"]
py = arrays['muon']["MCParticles.momentum.y"]
pz = arrays['muon']["MCParticles.momentum.z"]
mc_theta = np.degrees(np.arccos(pz/np.sqrt(px**2+py**2+pz**2)))
mc_phi = np.degrees(np.arctan2(py,px))
pid = arrays['muon']["MCParticles.PDG"]
plt.hist(ak.flatten(mc_theta[pid==13]),label=r"$\theta$", bins = np.linspace(-180,180,361))
plt.hist(ak.flatten(mc_phi[pid==13]),label=r"$\phi$", bins = np.linspace(-180,180,361))
#plt.xlim(-180,360)
plt.xlabel("Angle (Degrees)")
plt.legend()
plt.title("MC Particle Energy Comparison")
```

```python
cut = [len(row)==1 for row in pid]
output = plt.hist(ak.flatten(arrays['muon']["HcalFarForwardZDCHits.energy"][cut])*1000,histtype='step',bins=100)
bin_center = output[1][1:]/2+output[1][:-1]/2
coeff, covar = curve_fit(gaus,bin_center,output[0])
plt.plot(np.linspace(0.5,2,50),gaus(np.linspace(0.5,2,50),*coeff),label=rf'Muon: $\mu$ = {coeff[1]*1000:.2f} $\pm$ {np.abs(coeff[2])*1000:.2f} MeV')
MIP = coeff[1]

plt.xlabel('Energy per Cell (MeV)')
#plt.xlim(0,3)
plt.title('Muon: 4 GeV\nPosition: 0, 1m, 0.25m, Direction: 0, -1, 0')
plt.legend()
```

```python
MIP_cut = arrays['muon']["HcalFarForwardZDCHits.energy"] > 0.5*MIP/1000
lengths = [len(row) for row in arrays['muon']["HcalFarForwardZDCHits.energy"][MIP_cut]]
plt.hist(lengths,histtype='step',bins=np.linspace(1,8,8),label='Muon + Daughters')
cut = [len(row)==1 for row in arrays['muon']["MCParticles.PDG"]]
plt.hist(np.array(lengths)[cut],histtype='step',bins=np.linspace(1,8,8),label='Only Muon')
#plt.yscale('log')
plt.xlabel('Hits per Event')
plt.ylabel('Counts')
plt.title('Muon: 4 GeV, Cut: 0.5*MIP')
plt.legend()
```

```python
plt.hist(ak.flatten(mc_theta[pid==13]),label=r"Truth $\theta$", bins = 200,histtype='step')
plt.hist(ak.flatten(mc_phi[pid==13]),label=r"Truth $\phi$", bins = 200,histtype='step')

MIP_cut = arrays['muon']["HcalFarForwardZDCHits.energy"] > 0.5*MIP/1000
cut = [True if (len(cells)==2 and len(pid)==1) else False for cells, pid in zip(arrays['muon']["HcalFarForwardZDCHits.energy"][MIP_cut],arrays['muon']["MCParticles.PDG"])]
x = arrays['muon']["HcalFarForwardZDCHits.position.x"][np.array(cut)]
y = arrays['muon']["HcalFarForwardZDCHits.position.y"][np.array(cut)]
z = arrays['muon']["HcalFarForwardZDCHits.position.z"][np.array(cut)]

x_vec = x[:,1] - x[:,0]
y_vec = y[:,1] - y[:,0]
z_vec = z[:,1] - z[:,0]

theta = np.degrees(np.arccos(z_vec/np.sqrt(x_vec**2+y_vec**2+z_vec**2)))
output = plt.hist(theta, bins = np.linspace(-180,180,361),histtype='step',color='b',label=r"Reconstructed $\theta$")
#bin_center = output[1][1:]/2+output[1][:-1]/2
#coeff, covar = curve_fit(gaus,bin_center,output[0])
#plt.plot(np.linspace(-25,25,50),gaus(np.linspace(-25,25,50),*coeff),label='Theta Unweighted:\n'+ rf'$\mu$ = {coeff[1]:.2f}; $\sigma$ = {coeff[2]:.2f}',c='b')

phi = np.degrees(np.arctan2(y_vec, x_vec))
output = plt.hist(phi, bins = np.linspace(-180,180,361),histtype='step',color='r',label=r"Reconstructed $\phi$")
#bin_center = output[1][1:]/2+output[1][:-1]/2
#coeff, covar = curve_fit(gaus,bin_center,output[0])
#plt.plot(np.linspace(-25,25,50),gaus(np.linspace(-25,25,50),*coeff),label='Phi Unweighted:\n'+rf'$\mu$ = {coeff[1]:.2f}; $\sigma$ = {coeff[2]:.2f}',c='r')



plt.xlabel(r'Angle (degrees)')
plt.ylabel('Counts')
plt.title('Muon: 4 GeV, Hits per Events: 2, Cut: 0.5*MIP\n '+r'Theta: 90$\degree$-150$\degree$, Phi: 0$\degree$-360$\degree$')
#plt.yscale('log')
plt.legend()
```

```python
def theta_reconstruct(x, y, z):
    return np.degrees(np.arccos(z/np.sqrt(x**2+y**2+z**2)))

def phi_reconstruct(x, y, z):
    return np.degrees(np.arctan2(y, x))

def vector_angle_reconstruct(x, y, z):
    x_vec = x[:,1] - x[:,0]
    y_vec = y[:,1] - y[:,0]
    z_vec = z[:,1] - z[:,0]

    theta = theta_reconstruct(x_vec, y_vec, z_vec)
    phi = phi_reconstruct(x_vec, y_vec, z_vec)

    return theta, phi
    

cut = [True if (len(cells)==2 and len(pid)==1) else False for cells, pid in zip(arrays['muon']["HcalFarForwardZDCHits.energy"][MIP_cut],arrays['muon']["MCParticles.PDG"])]

px = arrays['muon']["MCParticles.momentum.x"]
py = arrays['muon']["MCParticles.momentum.y"]
pz = arrays['muon']["MCParticles.momentum.z"]
mc_theta = theta_reconstruct(px, py, pz)
mc_phi = phi_reconstruct(px, py, pz)


x = arrays['muon']["HcalFarForwardZDCHits.position.x"][np.array(cut)]
y = arrays['muon']["HcalFarForwardZDCHits.position.y"][np.array(cut)]
z = arrays['muon']["HcalFarForwardZDCHits.position.z"][np.array(cut)]

theta, phi = vector_angle_reconstruct(x, y, z)
```

```python
counts, bins = np.histogram(theta-ak.flatten(mc_theta[np.array(cut)]), bins = 100)
bin_center = bins[1:]/2+bins[:-1]/2
coeff, covar = curve_fit(gaus,np.array(bin_center),np.array(counts))
plt.errorbar(bin_center,counts,fmt='-o',label=rf'$\Delta$ Theta; $\sigma$ = {coeff[2]:.2f}$\degree$', c='b')
plt.errorbar(np.linspace(-20,20,100),gaus(np.linspace(-20,20,100),*coeff),fmt='--', c='b')

counts, bins = np.histogram(phi-ak.flatten(mc_phi[np.array(cut)]), bins = 100)
bin_center = bins[1:]/2+bins[:-1]/2
coeff, covar = curve_fit(gaus,np.array(bin_center),np.array(counts))
plt.errorbar(bin_center,counts,fmt='-o',label=rf'$\Delta$ Phi; $\sigma$ = {coeff[2]:.2f}$\degree$', c='r')
plt.errorbar(np.linspace(-20,20,100),gaus(np.linspace(-20,20,100),*coeff),fmt='--', c='r')

plt.xlabel(r'Reconstructed - MC Particle(degrees)')
plt.ylabel('Counts')
plt.title('Muon: 4 GeV, Hits per Events: 2, Cut: 0.5*MIP\n'+r'Theta: 65$\degree$-150$\degree$, Phi: 0$\degree$-360$\degree$'+'\nWith respect to Hexagon Detector frame')
plt.xlim(-50,50)
plt.legend()
```
